```python
from sklearn.preprocessing import scale
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

# Normalisation de données

## creations du matrice X :

```python
X = np.array([[1, -1, 2],[2, 0, 0],[0, 1, -1]])
```

## affichage du matrice X :

```python
print(X)
```

## le moyenne et  le variance du X :

```python
mean = X.mean()
varience = X.var()
print('mean is : ',mean,'and variance of matrix is : ',varience)
```

## normalisation du matrice X

```python
print('X matrix ==> : \n ',X)
scal_X = scale( X, axis=0, with_mean=True, with_std=True, copy=True )
print('normalized matrix ==> : \n',scal_X)
```

## le moyenne et  le variance du matrice normaliser :

```python
print('mean is : ',scal_X.mean(),'and variance of matrix is : ',scal_X.var())
```

# Normalisation MinMax

## creations du matrice X2 :

```python
X2 = np.array([[1, -1, 2],[2, 0, 0],[0, 1, -1]])
```

## affichage du matrice X2 :

```python
print(X2)
```

## la moyenne sur les variables

```python
X2.mean(0)
```

## Normalisation les données dans l’intervalle [0 1].

```python
scal2 = MinMaxScaler((0,1))
X2_norm = scal2.fit_transform(X2)
```

## affichage de matrice

```python
print(X2_norm)
```

## la moyenne sur les variables du matrice normaliser

```python
print('mean is : ',X2_norm.mean(0))
```

# visualisation de données

```python
iris = datasets.load_iris()
features = iris.data
targets = iris.target
print(features[:,1])
```

```python
plt.subplots(4,4,figsize=(20,10))
cmp = 1
for i in range(4):
    for j in range(4):
        plt.subplot(4,4,cmp)
#         print(cmp,i,j)
        plt.scatter(features[:,i], features[:,j], c=targets)
        plt.title('first generalized data')
        plt.xlabel('Variable x1')
        plt.ylabel('Variable x2')
        cmp=cmp+1

plt.show()
```

# Réduction de dimensions et visualisation de données

```python
pca = PCA(2)
lda = LDA()
iris_pca = pca.fit_transform(features)
iris_lda = lda.fit_transform(features,targets)
```

```python
plt.subplots(1,2,figsize=(20,10))

plt.subplot(1,2,1)
plt.scatter(iris_pca[:,0], iris_pca[:,1], c=targets)
plt.title('first generalized data')
plt.xlabel('Variable x1')
plt.ylabel('Variable x2')

plt.subplot(1,2,2)
plt.scatter(iris_lda[:,0], iris_lda[:,1], c=targets)
plt.title('first generalized data')
plt.xlabel('Variable x1')
plt.ylabel('Variable x2')

plt.show()
```
